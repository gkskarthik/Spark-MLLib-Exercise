## Loading the file into dataframe

```python
## LOADING FILES AND DISPLAYING THE TABLE
# File location and type
file_location = "/FileStore/tables/creditcard.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
```

## RandomForest Classifier

```python
RANDOM_SEED = 13579
TRAINING_DATA_RATIO = 0.7

RF_NUM_TREES = 3
RF_MAX_DEPTH = 4
RF_MAX_BINS = 32
```

```python
## Data Preprocessing and Test Train Split
from pyspark.mllib.linalg import Vectors
from pyspark.mllib.regression import LabeledPoint

transformed_df = df.rdd.map(lambda row: LabeledPoint(row[-1], Vectors.dense(row[0:-1])))

splits = [TRAINING_DATA_RATIO, 1.0 - TRAINING_DATA_RATIO]
training_data, test_data = transformed_df.randomSplit(splits, RANDOM_SEED)

print("Number of training set rows: %d" % training_data.count())
print("Number of test set rows: %d" % test_data.count())
```

```python
## Training RandomForest Model
from pyspark.mllib.tree import RandomForest

model_rf = RandomForest.trainClassifier(training_data, numClasses=2, categoricalFeaturesInfo={}, \
    numTrees=RF_NUM_TREES, featureSubsetStrategy="auto", impurity="gini", \
    maxDepth=RF_MAX_DEPTH, maxBins=RF_MAX_BINS, seed=RANDOM_SEED)
```

```python
## Calculating the accuracy for the RandomForest Model
predictions_rf = model_rf.predict(test_data.map(lambda x: x.features))
labels_and_predictions_rf = test_data.map(lambda x: x.label).zip(predictions_rf)
acc_rf = labels_and_predictions_rf.filter(lambda x: x[0] == x[1]).count() / float(test_data.count())
print("Model accuracy: %.3f%%" % (acc_rf * 100))
```

```python
from pyspark.mllib.evaluation import MulticlassMetrics

metrics_rf = MulticlassMetrics(labels_and_predictions_rf)
precision_rf = metrics_rf.precision()
recall_rf = metrics_rf.recall()
fMeasure_rf = metrics_rf.fMeasure()
print("Model precision: %f, recall: %f and f1 score:%f " %((precision_rf* 100), (recall_rf* 100), (fMeasure_rf* 100)))
```

```python
## Evaluating the metrics for the RandomForest Model
from pyspark.mllib.evaluation import BinaryClassificationMetrics

metrics_rf = BinaryClassificationMetrics(labels_and_predictions_rf)
print("Area under Precision/Recall (PR) curve: %.f" % (metrics_rf.areaUnderPR * 100))
print("Area under Receiver Operating Characteristic (ROC) curve: %.3f" % (metrics_rf.areaUnderROC * 100))
```

### MultiLayered Perceptron Classifier

```python
from pyspark.ml.linalg import DenseVector

input_data = df.rdd.map(lambda x: (x[-1], DenseVector(x[0:-1])))

df = spark.createDataFrame(input_data, ["label", "features"])
```

```python
training_data1, test_data1 = df.randomSplit([.8,.2],seed=1234)

print("Number of training set rows: %d" % training_data1.count())
print("Number of test set rows: %d" % test_data1.count())
```

```python
from pyspark.ml.classification import MultilayerPerceptronClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

layers = [30, 5, 4, 2]
trainer = MultilayerPerceptronClassifier(maxIter=100, layers=layers, blockSize=128, seed=1234)

model_p = trainer.fit(training_data1)
```

```python
result_p = model_p.transform(test_data1)
predictionAndLabels = result_p.select("prediction", "label")
evaluator = MulticlassClassificationEvaluator(metricName="accuracy")

print("Test set accuracy = " + str(evaluator.evaluate(predictionAndLabels) * 100))
```

```python
precision_p = MulticlassClassificationEvaluator(metricName="weightedPrecision")
recall_p = MulticlassClassificationEvaluator(metricName="weightedRecall")
fScore_p = MulticlassClassificationEvaluator(metricName="f1")

print("Precision = " + str(precision_p.evaluate(predictionAndLabels) * 100))
print("Recall = " + str(recall_p.evaluate(predictionAndLabels) * 100))
print("f1 Score = " + str(fScore_p.evaluate(predictionAndLabels) * 100))
```
